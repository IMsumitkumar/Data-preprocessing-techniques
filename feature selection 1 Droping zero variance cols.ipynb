```python
data = pd.read_csv("/home/nucleya/Desktop/preprocessed_titanic.csv")

data = data.drop(columns='Unnamed: 0')
data.head()
```

```python
# removing zero as well as and near zero variance


# count of unique points divided by len(feature) has to be lower than a prespecified threshold
# count of most common point divided by second most common point has to be greater than a prespecified threshold

def remove_zero_nearZero(data, target):
    to_drop = []
    for i in data.drop(target, axis=1).columns:
        fu = pd.DataFrame(data[i].value_counts()).sort_values(by=i, ascending=False, inplace=False)
        for_zero = len(fu)/len(data[target])
        if len(fu[i]) == 1:
            for_near_zero = 100
        else:
            for_near_zero = fu.iloc[0, 0] / fu.iloc[1, 0]
            
        if (for_zero <= 0.1) and (for_near_zero >= 20):
            to_drop.append(i)
            
    data.drop(to_drop, axis=1, inplace=True)
        
    print(to_drop)
    return data
            
    
```

```python
remove_zero_nearZero(data, 'Survived')
```
